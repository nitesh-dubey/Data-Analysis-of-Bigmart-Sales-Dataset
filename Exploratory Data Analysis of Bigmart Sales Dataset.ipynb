Now Let us Load the 'Bigmart Sales Dataset' which Depicts various attributes regarding the sales of many products at 10 stores across different cities and view its first few rows

```python
data = pd.read_csv('bigmart_data.csv')
data.head()
```

We see that there are missing values in dataset, so let us drop those rows and reset index

```python
data.dropna(axis = 0, how = 'any',inplace = True)
data.reset_index(drop = True)
data.head()
```

Let us look how the 'Item_Outlet Sales' and 'Item_MRP' vary with Item Type 

# Item Type VS Item MRP

```python
price_by_item = data.groupby('Item_Type').Item_MRP.mean()
price_by_item
```

```python
plt.style.use('ggplot')
x = price_by_item.index.tolist()
y = price_by_item.values.tolist()
plt.figure(figsize = (15,7))
plt.xlabel('Item Type')
plt.ylabel('Item MRP')
plt.title('Item Type VS Item MRP')
plt.xticks(ticks = np.arange(len(x)), labels = x, rotation = 45)
plt.plot(x,y,'k-')
```

By Plotting Line chart between Item Type and Item MRP we observe following facts

        1.) The Mean price of Starchy Foods leads the group, which is closely followed by the Price of Household item
            dairy products
            
        2.) Baking Goods, health and hygiene products and softdrinks have low MRP

# Item_Type Vs Item_outlet_sales

```python
sales_by_item = data.groupby('Item_Type').Item_Outlet_Sales.mean()
sales_by_item
```

```python
plt.style.use('seaborn')
x = sales_by_item.index.tolist()
y = sales_by_item.values.tolist()
plt.figure(figsize = (15,7))
plt.xlabel('Item Type')
plt.ylabel('Item Outlet Sales')
plt.title('Item Type VS Item Outlet Sales')
plt.xticks(ticks = np.arange(len(x)), labels = x, rotation = 45)
plt.plot(x,y)
```

Comparing The above 2 Plot shows us very Interesting feature..

1.)Although The price of Starchy Food is highest but it clearly doesnt dominate Outlet_Sales...

2.)We Can clearly observe that the Outlet_sales of Seafood is highest , eventhough its price is not highest which shows 
   shows that it is a profitable business
   
3.)Baking Goods trade will obviously at loss because even though it has lowest price, its sales is lowest.

# Outlet Establishment Year vs Item outlet sales

```python
price_by_year = data.groupby('Outlet_Establishment_Year').Item_Outlet_Sales.mean()
price_by_year.sort_index(inplace = True)
```

```python
plt.style.use('ggplot')
plt.figure(figsize = (15,8))
x = price_by_year.index.map(lambda x : str(x)).tolist()
y = price_by_year.values.tolist()
plt.xlabel('Outlet_Establishment_Year')
plt.ylabel('Item_Outlet_Sales')
plt.title('Outlet_Establishment_Year vs Item_Outlet_Sales')
plt.xticks(ticks = np.arange(len(x)), labels = x)
plt.plot(x,y,'r')
```

The above figure proves our assumption that old is Gold!!! 

The outlets which are Established Earlier than have greater outlet sales than the outlets which were established later.

Although The outlets which are very old have lost some of their shine and have a little lesser sales than the outlets which are a some years newer than them. but obviously have greater sales than very new outlets.

# Method To Prepare Data For Machine Learning Algorithm

We have already Dropped the rows with missing values. we could also have replaced the missing value with the mean value of that attribute. 

In this dataset we can observe that many attributes are categorical variables. So In order to Prepare This dataset for applying machine learning algorithm, we need to convert the catagorical variables into numbers, as machine learning algorithms will only act on numbers.

For Example ::

In Outlet_Size attribute, we can change  High to 3,  Medium to 2,  Small to 1, 
In Outlet_Location_Type attribute, we can change  Tier 1 to 1,  Tier 2 to 2,  Tier 3 to 3

In Below Example it is depicted how to convert catagorical variables to numbers for 'Outlet_Size' attribute..
The Other attributes will follow the same method.

```python
data.head()
```

```python
def preparedata(x):
    if x == 'High':
        return 3
    elif x == 'Medium':
        return 2
    elif x == 'Small':
        return 1
```

```python
newdata = data
newdata['Outlet_Size'] = newdata['Outlet_Size'].apply(preparedata)
```
